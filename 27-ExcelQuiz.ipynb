```python
#4) Bu firmada title (senior - junior) durumuna göre ortalama maaş karşılaştırması nasıldır?
```

```python
dataFrame_title = dataFrame["Title"]
SalaryforTitle = dataFrame["Salary"]
yeni_dataFrame = pd.concat([dataFrame_title, SalaryforTitle], axis=1)
yeni_dataFrame.groupby("Title").mean()
```

```python
#5) Senior bir kişinin junior bir kişiye göre maaşı ortalama yüzde kaç fazladır?
```

```python
senior = yeni_dataFrame[yeni_dataFrame["Title"] == "Senior"]
junior = yeni_dataFrame[yeni_dataFrame["Title"] == "Junior"]
senior_mean, junior_mean = senior["Salary"].mean(), junior["Salary"].mean()
fark = ((senior_mean - junior_mean) / junior_mean) * 100
print(f"Senior mean: {round(senior_mean,3)}\nJunior mean: {round(junior_mean,3)}\nFark: {round(fark,3)}")
```

```python
#6) Software development departmanında senior bir kişinin junior bir kişiye göre maaşı ortalama ne kadar fazladır?
```

```python
Softfware_Department_data = dataFrame[dataFrame["Department"] == "Software Development"]
senior, junior = Softfware_Department_data[Softfware_Department_data["Title"] == "Senior"], Softfware_Department_data[Softfware_Department_data["Title"] == "Junior"]
senior_mean, junior_mean = senior["Salary"].mean(), junior["Salary"].mean()
fark = (senior_mean / junior_mean)
print(f"Senior mean: {round(senior_mean,3)}\nJunior mean: {round(junior_mean,3)}\nFark: {round(fark,3)}")
```

```python
#7) Finance departmanında c-level bir kişinin mid-senior bir kişiye göre maaşı ortalama ne kadar fazladır?
```

```python
Finance_Department_data = dataFrame[dataFrame["Department"] == "Finance"]
c_level, mid_senior = Finance_Department_data[Finance_Department_data["Title"] == "C-level"], Finance_Department_data[Finance_Department_data["Title"] == "Mid-Senior"]
c_level_mean, mid_senior_mean = c_level["Salary"].mean(), mid_senior["Salary"].mean()
fark = ((c_level_mean - mid_senior_mean) / mid_senior_mean) * 100
print(f"C-level mean: {round(c_level_mean,3)}\nMid-senior mean: {round(mid_senior_mean,3)}\nFark: {round(fark,3)}")
```

```python
#8) Software development departmanında c-level çalışan sayısı marketing departmanında çalışana oranla kaç kat fazladır?
```

```python
Softfware_Department_data = dataFrame[dataFrame["Department"] == "Software Development"]
Marketing_Department_data = dataFrame[dataFrame["Department"] == "Marketing"]
Softfware_Department_c_level = Softfware_Department_data[Softfware_Department_data["Title"] == "C-level"]
Marketing_Department_c_level = Marketing_Department_data[Marketing_Department_data["Title"] == "C-level"]
S_C_L_calisan_sayisi = Softfware_Department_c_level.shape[0]
MRKT_C_L_calisan_sayisi = Marketing_Department_c_level.shape[0]
katsayı = S_C_L_calisan_sayisi / MRKT_C_L_calisan_sayisi
print(f"Software Department C-level Çalışan Sayısı:{S_C_L_calisan_sayisi}\nMarketing Department C-level Çalışan Sayısı : {MRKT_C_L_calisan_sayisi}\nFark : {katsayı}")
```
